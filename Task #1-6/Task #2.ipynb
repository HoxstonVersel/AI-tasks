```python
#pandas это высокоуровневая Python библиотека для анализа данных.
#Библиотека Math в Python обеспечивает доступ к некоторым популярным математическим функциям
import pandas as pd
import math
column_a = 'Play'
column_b = 'Weather'
column_с = 'Field'
data = pd.DataFrame({column_a:['no', 'no', 'no', 'yes', 'yes', 'no', 'no','yes','yes','yes', 'yes','yes','yes','yes'],
                     column_b:['Rainy','Rainy','Rainy', 'Rainy', 'Rainy','Sunny','Sunny','Sunny','Sunny','Sunny', 'Overcast','Overcast','Overcast','Overcast'],
                     column_с:['bad','bad','bad', 'bad', 'good','good','good','good','bad','good', 'bad','good','bad','good']
                    })

B_hypothes = 'Sunny'
C_hypothes = 'good'
A_value_yes = 'yes'
A_value_no = 'no'

data_only_a_yes=data[data[column_a] == A_value_yes]
print('data_only_a_yes=\n',data_only_a_yes)

data_only_a_no=data[data[column_a] == A_value_no]
print('data_only_a_no=\n',data_only_a_no)

data_h_b_yes=data_only_a_yes[data_only_a_yes[column_b] == B_hypothes]
data_h_c_yes=data_only_a_yes[data_only_a_yes[column_с] == C_hypothes]
print('data_h_b_yes=\n',data_h_b_yes)
print('data_h_c_yes=\n',data_h_c_yes)

data_h_b_no=data_only_a_no[data_only_a_no[column_b] == B_hypothes]
data_h_c_no=data_only_a_no[data_only_a_no[column_с] == C_hypothes]
print('data_h_b_no=\n',data_h_b_no)
print('data_h_c_no=\n',data_h_c_no)

b_yes=len(data_h_b_yes)/len(data_only_a_yes) 
print('b_yes=',b_yes)

b_no=len(data_h_b_no)/len(data_only_a_no)
print('b_no=',b_no)

c_yes = len(data_h_c_yes)/len(data_only_a_yes)
print('c_yes=',c_yes)

c_no = len(data_h_c_no)/len(data_only_a_no)
print('c_no=',c_no)

p_yes=(b_yes/b_no) * (c_yes/c_no)

p_yes
```

```python
#Laplacian smoothing
## УНИКАЛЬНЫЕ ЗНАЧЕНИЯ - YES AND NO - 2 
# 
B_hypothes = 'Storm'
C_hypothes = 'bad'
A_value_yes = 'yes'
A_value_no = 'no'

data_only_a_yes = data[data[column_a] == A_value_yes]
print('data_only_a_yes=\n',data_only_a_yes)

data_only_a_no = data[data[column_a] == A_value_no]
print('data_only_a_no=\n',data_only_a_no)

data_h_b_yes = data_only_a_yes[data_only_a_yes[column_b] == B_hypothes]
data_h_c_yes = data_only_a_yes[data_only_a_yes[column_с] == C_hypothes]
print('data_h_b_yes=\n',data_h_b_yes)
print('data_h_c_yes=\n',data_h_c_yes)

data_h_b_no = data_only_a_no[data_only_a_no[column_b] == B_hypothes]
data_h_c_no = data_only_a_no[data_only_a_no[column_с] == C_hypothes]
print('data_h_b_no=\n',data_h_b_no)
print('data_h_c_no=\n',data_h_c_no)

#Cглаживание лапласа
b_yes=(len(data_h_b_yes)+1)/(len(data_only_a_yes)+2)
print('b_yes=',b_yes)

b_no=(len(data_h_b_no)+1)/(len(data_only_a_no)+2)
print('b_no=',b_no)

c_yes = (len(data_h_c_yes)+1)/(len(data_only_a_yes)+2)
print('c_yes=',c_yes)

c_no = (len(data_h_c_no)+1)/(len(data_only_a_no)+2)
print('c_no=',c_no)

p_yes=(b_yes/b_no) * (c_yes/c_no)

p_yes
```
